Aides :

L'approche graphique de ce type d'exercice (modélisation par une fonction linéaire) est traitée dans la partie <b>"tracer un graphique"</b> avec le bonus <b>"curseurs"</b>. Vous pouvez effectuer un copier - coller de la partie utile puis adapter les noms des variables.

```python
# écrire ici votre programme :
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import linregress

# Avec des listes pré-remplies en degrés :
ideg = [15,25,35,45,55]
rdeg = [10,17,23,29,34]

print ("Quelle est la valeur de l'indice du milieu 1 ?")
n1 = float(input("n1 = "))

# Conversion des angles donnés en degré vers radian
irad = [(VAL*np.pi/180) for VAL in ideg]
rrad = [(VAL*np.pi/180) for VAL in rdeg] 

####### Autre façon de faire : #######
##irad = []
##rrad = []
##for VAL in ideg:             
##    irad.append(VAL*np.pi/180)
##for VAL in rdeg:             
##    irad.append(VAL*np.pi/180)
#######


# Calcul de l'indice n2
n2 = []
for x in range (len(irad)):            # Les 2 listes irad et rrad ont le même nombre d'éléments
                                       # On calcule la valeur n pour chaque couple (irad;rrad) 
                                       # et on la met à la fin de la liste des n2.
    n = [n1*np.sin(irad[x])/np.sin(rrad[x])]
    n2.append(n)
# print(n2)

# Calcul de la moyenne des valeurs de n2
moy = np.mean(n2)                         # La fonction mean calcule la moyenne de svaleurs contenue dans n2.
print("La valeur moyenne pour l'indice n2 est :",round(moy,2))

############################################################################################
############ Vous pouvez aussi déterminer cette valeur par régression linéaire.#############
############################################################################################

# L'abscisse correspond  à sin(irad) et l'ordonnée à sin(rrad)
X = [np.sin(VAL) for VAL in irad]
Y = [np.sin(VAL) for VAL in rrad]

# détermination des coefficients      
a,b,ro,_,_= linregress(X,Y)  # Régression linéaire 

# Abscisses et ordonnées du modèle
X_modele = np.linspace(0,max(X),101)          #permet de créer une liste de 101 valeurs d'abscisses
Y_modele = [a*i+b for i in X_modele]          #calcul l'ordonnée en appliquant le modèle

# tracé du graphique
plt.plot(X,Y,'r+')              # les couples de valeurs sont représentées par des croix rouges
plt.plot(X_modele,Y_modele)     # par défaut les points sont liés par un segment de droite bleu

# Informations sur le graphique
plt.xlabel('sin(i) en rad')        # légende de l'axe des absisses
plt.ylabel('sin(r) en rad')        # légende de l'axe des ordonnées
plt.legend(('sin(i) = f(sin(r))','modèle'))   # légende des courbes sur le graphique

# Montrer le graphique
plt.show()

#Affichage des résultats
print("Coefficient directeur : a =",a)
print("Ordonnée à l'origine : b =",b)
print("Coefficient de corrélation : r = ",ro)

# Calcul de la valeur de n2 = n1 / a
n2 = n1 / a
print("La valeur de n2 obtenue par régression linéaire vaut :",round(n2,2))
```

    <center><b> En quoi utilise - t - on ici une boucle "while" ? </b></center><br> Le graphique tracera les quantités de matière pour des abscisses comprises entre zéro et x<sub>max</sub> qui constitue donc une condition d'arrêt à la valeur de l'avancement et au de tracé de la courbe.

<div class="alert alert-warning">
   Dans la situation suivante, le diiode est le réactif limitant et la valeur de l'avancement maximum est x<sub>max</sub> = 0,05 mol.<br><b> Pour construire l'évolution des quantités de matière des réactifs et produits du système, il faut faire varier la valeur de l'avancement depuis zéro jusqu'à x<sub>max</sub>.</b> Cette valeur <b>x</b> avance suivant un pas qui est choisi suffisamment petit pour que le tracé de la courbe soit propre.

```python
import matplotlib.pyplot as plt

# Prise en compte de l'équation de réaction du type : aA + bB -> cC 
a = 2               # Valeur qui pourrait être demandée puis input dans la variable a sous format d'un entier
b = 3
c = 2
nA = 0.20           # qté de matière initiale pour le réactif A
nB = 0.15           # qté de matière initiale pour le réactif B
nC = 0              # qté de matière initiale pour le produit C

# Gestion de l'avancement
Xmax = 0.05         # Valeur de l'avancement maximum en mol
x = 0               # initialisation de l'avancement à zéro, cette valeur va évoluer jusqu'à Xmax
dx = 0.001          # incrément d'avancement qui permet de passer au point suivant
LX = []             # Liste qui stocke les valeurs successives de l'avancement

# Listes qui vont stocker les quantités de matière des réactif et du produit
LNA = []          # pour le réactif A
LNB = []          # pour le réactif B
LNC = []          # pour le produit C

while x<=Xmax :
    LX.append(x)    # on complète la liste LX par la valeur calculée x
    LNA.append(nA-a*x)   # de même pour la liste NA complétée par la valeur calculée nA-a*x
    LNB.append(nB-b*x)   # idem pour les autres
    LNC.append(nC+c*x)
    x=x+dx          # on remplace x par x+dx pour passer au point suivant du graphique

plt.plot(LX,LNA,'r-',label='nA')  # on fait afficher le graphe, avec les options classiques de tracé
plt.plot(LX,LNB,'g-',label='nB')  # idem pour les autres
plt.plot(LX,LNC,'b-',label='nC')
plt.grid()
plt.xlabel('x (mol)')
plt.ylabel('n (mol)')
plt.legend()
plt.show()

```

```python
# rattachement des librairies gérant les tracés de courbes et certains outils mathématiques 
import matplotlib.pyplot as plt
import numpy as np

#Nombres stoechiométriques
a = 1
b = 2
c = 1

# Affichage de l'équation de la réaction
print("l'équation étudiée est du type : ",
      a," A    +    ",b," B    -->     ",
      c," C ")

# Quantités de matières initiales
nA_0 = 2   
nB_0 = 2.6
nC_0 = 0

############Initialisation des variables############

# Initialisation de la chaine de caractère correspondant au réactif limitant
Rlimitant = ""  

# Avancement initial
x = 0            
# Pas de l'avancement (on augmentera progressivement x de la valeur dx)
dx = 0.001       

# Création des listes contenant les quantités de matière et initialisation de ces listes avec la valeur initiale  
nA = [2] 
nB = [2.6] 
nC = [0]

# Création et initialisation de la liste contenant l'avancement
X = [x]  


############ Calculs des quantités de matière en cours d'avancement############

# Ici, on considère que le système est en réaction tant qu'aucune quantité de matière des réactifs n'est nulle.
# On peut ainsi considérer que l'avancement évolue tant que les quantités de matière dernièrement rentrées 
#sont positives.

while nA[-1] > 0 and nB[-1] > 0 :   # L'indice -1 permet d'accéder au dernier l'élément de la liste
    x = x + dx
    X.append(x)
    nA.append(nA_0-a*x)
    nB.append(nB_0-b*x)
    nC.append(nC_0+c*x)        

#Détermination du réactif limitant 
if nA[-1]  <= 0                    : Rlimitant = "A"
if nB[-1]  <= 0                    : Rlimitant = "B"
if nA[-1]  <= 0   and   nB[-1]<= 0 : 
    Rlimitant="A et B : le mélange est stoechiometrique"
        
#Affichage des résultats
print('Le réactif limitant est ',Rlimitant,
      '\n Avancement maximum : ','{0:.4f}'.format(x),
      'mol' )
#{0:.4f} permet d’afficher un nombre arrondi à 4 chiffres après la virgule (ici).
```

**Affichage des courbes permettant de suivre l'évolution des quantités de matière**

La ligne de code 2 ci-dessous permet d'afficher le graphe de l'évolution de la quantité de matière de A en fonction de l'avancement X.<br>

```python
plt.figure(figsize=(10,10))
plt.plot(X,nA,'g-',lw=1,label='nA')
plt.plot(X,nB,'b-',lw=1,label='nB')    
plt.plot(X,nC,'r-',lw=1,label='nC')       
plt.grid()
plt.xlabel('x (mol)')
plt.ylabel('n (mol)')
plt.legend()
plt.show()
```
